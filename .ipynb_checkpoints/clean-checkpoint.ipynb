```python
import pandas as pd
import numpy as np

print("library load successful !!!")
```

```python
df = pd.read_csv(r"messy_ice_detention.csv", encoding="latin1")
```

```python
print(f"Rows: {df.shape[0]} Columns: {df.shape[1]}")
```

```python
df.head()
```

```python
df["Name"] = df["Name"].str.replace(r"[^A-Za-z0-9 ]", "", regex=True)
df
```

```python
df["Name"]=df["Name"].str.replace(r"\s+", " ", regex = True).str.strip()
```

```python
df["Name"]=df["Name"].str.title()
```

```python
df.head(10)
```

```python
df["City"]=df["City"].str.strip().str.title()
```

```python
df[df["City"].isna() | (df["City"] == "")]
```

```python
#The Geauga County Jail in Ohio is located in Chardon.
```

```python
df.at[51, "City"] = "Chardon"
```

```python
df.iloc[[51]]
```

```python
df[df["City"].isnull()]
```

```python
df[(df["State"].isnull()) | (df["State"] == " ")]
```

```python
city_to_state = {
    "Atlanta": "GA",
    "Encinal": "TX"
}

df["State"] = df["City"].map(city_to_state).fillna(df["State"])
# creating a dictionary to fill the states based on the City, .map applied on df["City"] to fill corresponding states, fillna(df["State"]) used to keep the already existing states as it is.
```

```python
df.iloc[["6","71"]]
```

```python
df["State"] = df["State"].str.strip()
```

```python
df.head()
```

```python
df = df.rename(columns={ "Last Inspection End Date": "Last_Inspected_Date"})
```

```python
df["Last_Inspected_Date"] = pd.to_numeric(df["Last_Inspected_Date"], errors = "coerce")
```

```python
np.dtype(df["Last_Inspected_Date"])
```

```python
df["Last_Inspected_Date"] = pd.to_datetime(df["Last_Inspected_Date"], origin = '1899-12-30', unit ='D', errors = "coerce")
```

```python
df.info()
```

```python
df.to_csv("Cleaned_file.csv")
```
